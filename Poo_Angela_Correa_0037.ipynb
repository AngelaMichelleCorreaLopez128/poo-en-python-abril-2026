```python
class Animal:
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    def ladrar(self):
        print(f"{self.nombre} de raza {self.raza} está ladrando: ¡Guau, guau! 🐱‍💻")

    def correr(self):
        print(f"{self.nombre} de raza {self.raza} está corriendo rápidamente.🐱‍🏍")

    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina.🐱‍👓")

# Creación de los objetos
pitbull = Animal("Rex", "Pitbull")
chihuahua = Animal("Taco", "Chihuahua")
# creamos el objeto perro callejero con datos
perro_callejero = Animal("Fido", "Chafa")
# Ejemplo de uso de las funciones
pitbull.ladrar()
chihuahua.correr()
perro_callejero.morder()


print("programa creado por Angela Correa NL = 13")
```

```python
class Animal:
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    def ladrar(self):
        print(f"{self.nombre} (Raza: {self.raza}) dice: ¡Guau, guau!")

    def correr(self):
        print(f"{self.nombre} (Raza: {self.raza}) está corriendo.")

    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina.🐱‍👓")

# Creación de los objetos identificados como perro1 y perro2
perro1 = Animal("Rocky", "Pitbull")
perro2 = Animal("Chispita", "Chihuahua")

# Ejecución de las funciones
print("--- Datos del Perro 1 ---")
perro1.ladrar()
perro1.correr()

print("\n--- Datos del Perro 2 ---")
perro2.ladrar()
perro2.correr()

# creamos el objeto perro callejero con datos
perro_callejero = Animal("Fido", "Chafa")
# Ejemplo de uso de las funciones
perro1.ladrar()
perro2.correr()
perro_callejero.morder()
print("programa creado por Angela Correa NL = 13")
```
